```python
# prompt: give an example of Least Squares

import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a column of ones to X for the intercept term
X_b = np.c_[np.ones((100, 1)), X]

# Calculate the optimal theta using the Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("Optimal theta:", theta_best)

# Make predictions using the calculated theta
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)

print("Predictions for X_new:", y_predict)

# Plot the original data and the fitted line
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.title("Least Squares Regression")
plt.grid(True)
plt.show()
```

```python
# prompt: read the file Study_vs_Score_data.csv

import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('Study_vs_Score_data.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# prompt: show graph about data with input column 'Attendance_Hours' and output column 'Final_Marks'

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['Attendance_Hours'], df['Final_Marks'])
plt.xlabel('Attendance Hours')
plt.ylabel('Final Marks')
plt.title('Relationship between Attendance Hours and Final Marks')
plt.grid(True)
plt.show()
```

```python
# prompt: Create a linear regression model for the data

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a Linear Regression model
lin_reg = LinearRegression()

# Reshape the features (X) and target (y) data if necessary
# If X and y are already in the correct shape (e.g., (n_samples, n_features) and (n_samples,)),
# you can skip the reshape. Assuming your DataFrame has columns 'Attendance_Hours' and 'Final_Marks'.
X_df = df[['Attendance_Hours']] # Features should be a DataFrame or 2D array
y_df = df['Final_Marks']       # Target can be a Series or 1D array

# Fit the model to the data
lin_reg.fit(X_df, y_df)

# Print the intercept and coefficients
print("Intercept:", lin_reg.intercept_)
print("Coefficient (slope):", lin_reg.coef_)

# You can now make predictions with the trained model
# For example, predict the score for 10 hours of attendance
hours_to_predict = np.array([[10]])
predicted_score = lin_reg.predict(hours_to_predict)
print(f"Predicted score for 10 hours of attendance: {predicted_score[0]:.2f}")

# Plot the original data and the linear regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_df, y_df, label='Original Data')
plt.plot(X_df, lin_reg.predict(X_df), color='red', label='Linear Regression Line')
plt.xlabel('Attendance Hours')
plt.ylabel('Final Marks')
plt.title('Linear Regression on Attendance Hours vs. Final Marks')
plt.legend()
plt.grid(True)
plt.show()
```
